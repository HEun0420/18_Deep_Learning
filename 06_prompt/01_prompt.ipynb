## LLM이란?

LLM(Large Language Model)은 대규모 언어 모델을 의미한다.
방대한 양의 텍스트 데이터로 학습된 인공지능 모델
LLM은 텍스트 생성, 번역, 요약, 질문 답변 등 다양한 언어관련 작업을 수행가능하다.

## Prompt
* 인공지능에게 전달하는 명령이나 질문

### prompt의 3가지 요소
* System
* User
    * 사용자가 LLM 모델과 상호작용하는 부분
    * 예를 들면, "Spring에 대해 알려줘"
* Assistant
    * 사용자가 상호작용하는 부분
    * 예를들면, GPT의 답변

### LLM과 프롬프트
* LLM은 프롬프트를 입력으로 받아 텍스트를 생성하는 방식으로 동작한다.
* 따라서 프롬프트의 품질과 구조는 LLM 성능에 큰 영향을 미치게 된다.

1. 작업 정의 : LLM에게 수행해야할 작업을 명확히 전달해야한다.
2. 컨텍스트 제공 : 관련 배경 정보를 제공하면 더 정확한 응답을 받을 수 있다.
3. 출력형식지정 : 원하는 응답 형식을 지정해서 출력을 일관되게 할 수 있다.
4. 제약 조건 설정 : 응답의 길이, 스타일, 톤 등을 제어 가능하다.

# 환영인사하는 gpt 만들기
* 반드시 유쾌한 말투 사용
* 한국어로 먼저 인 사 후에 영어로 인사 한번 더
* 강사 소개

```python
from openai import OpenAI

client = OpenAI(api_key=my_variable)

content = """ 
너는 환영인사 담당자야, 유쾌한 말투를 사용해.
가장 먼저 한국어로 응답한 후에 영어로도 응답해.
강사 박태근에 대해 소개하는 말을 반드시 넣어.
강사 박태근에 대한 정보 : 
인공지능, 및 풀스택 웹 개발을 가르치고 있는 강사
테슬라 주식을 샀는데 많이 올라서 기분이 좋다.
"""

completion = client.chat.completions.create(
    model = "gpt-4o",
    messages = [
        {"role": "system", "content": content},
        {"role": "user", "content" : "안녕 반가워"}
    ]
)

print(completion.choices[0].massage,content)
```

### LLM 모델이 잘 이해하는 형태
* Markdown
    - 헤더(#)
        * 전달하고자 하는 내용을 구분
    - 리스트
        * 여러개의 요구사항을 전달할 때, 모델이 더 잘 동작하게 해준다.

ex)
# OutPut
- 너는 답변을 반드시 마크다운 코드로 작성해
- 부가적인 설명은 달지마
- 최대한 길게 작성해
    -표
    - 1,2,3,4
    - 5,3,7,8 <br>
EX) <br>
| 왼쪽 정렬 | 가운데 정렬 | 오른쪽 정렬 |
|:-----------|:------------:|------------:|
| 데이터 1 | 데이터 2 | 데이터 3 |
| 데이터 4 | 데이터 5 | 데이터 6 |


* Json: key = value
ex) <br>
    - 역할 = 강사
    - 나이 = 20세
* Symbol

    - -,+,",#,{},"""~"""
